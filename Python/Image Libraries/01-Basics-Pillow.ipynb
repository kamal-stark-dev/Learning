```python
from PIL import Image
import matplotlib.pyplot as plt
```

```python
# selecting an image
image = Image.open('images/billie.png')
```

```python
# show image
plt.axis("off")
plt.imshow(image)
```

```python
print(f'''
Image Format: {image.format}
Image Size: {image.size}
Image Mode: {image.mode}
''')
```

```python
from PIL import ImageOps # contains ready made image processing operations
```

```python
image_gray = ImageOps.grayscale(image)

plt.axis("off")
plt.imshow(image_gray, cmap="gray")
```

```python
image_gray.mode # 'L' means Luminance
```

```python
image_gray.save('images/billie-grayscale.jpg') # you can also change the type of image
```

```python
# We can use the quantize method to quantize the image
# image_gray.quantize(2) - (number of distinct colors or intensity levels of an image)

fig, axes = plt.subplots(1, 2, figsize=(10, 5))  # 1 row, 2 columns

axes[0].imshow(image_gray.quantize(2))
axes[0].set_title("Quantize = 2")
axes[0].axis("off")

axes[1].imshow(image_gray.quantize(6))
axes[1].set_title("Quantize = 6")
axes[1].axis("off")

plt.tight_layout()
plt.show()
```

```python
# we can obtain different red, green, blue channels

red, green, blue = image.split()

plt.axis("off")
plt.imshow(red)
```

```python
# Creating Colorized Version of Red
merged_red = Image.merge("RGB", (red, Image.new("L", red.size), Image.new("L", red.size)))

plt.axis("off")
plt.imshow(merged_red)
```

```python
red, green, blue = image.split()

# Create colorized versions
red_img   = Image.merge("RGB", (red, Image.new("L", red.size), Image.new("L", red.size)))
green_img = Image.merge("RGB", (Image.new("L", green.size), green, Image.new("L", green.size)))
blue_img  = Image.merge("RGB", (Image.new("L", blue.size), Image.new("L", blue.size), blue))

fig, axs = plt.subplots(2, 4, figsize=(16, 8))

# Original
axs[0,0].imshow(image)
axs[0,0].set_title("Original")
axs[0,0].axis("off")

# Grayscale channels
axs[0,1].imshow(red, cmap="gray")
axs[0,1].set_title("Red (Grayscale)")
axs[0,1].axis("off")

axs[0,2].imshow(green, cmap="gray")
axs[0,2].set_title("Green (Grayscale)")
axs[0,2].axis("off")

axs[0,3].imshow(blue, cmap="gray")
axs[0,3].set_title("Blue (Grayscale)")
axs[0,3].axis("off")

# Colorized channels
axs[1,1].imshow(red_img)
axs[1,1].set_title("Red (Colorized)")
axs[1,1].axis("off")

axs[1,2].imshow(green_img)
axs[1,2].set_title("Green (Colorized)")
axs[1,2].axis("off")

axs[1,3].imshow(blue_img)
axs[1,3].set_title("Blue (Colorized)")
axs[1,3].axis("off")

axs[1,0].axis("off")  # keep blank spot under Original

plt.tight_layout()
plt.show()
```
