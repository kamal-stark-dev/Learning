```python
from PIL import Image
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show_image(image):
    plt.imshow(image)
    plt.axis("off")
    plt.gca().set_position([0, 0, 1, 1])
    plt.show()
```

```python
elizabeth = cv2.imread("images/elizabeth-olsen.png")
elizabeth = cv2.cvtColor(elizabeth, cv2.COLOR_BGR2RGB)

show_image(elizabeth)
```

### **Copy Image**

```python
# if you make changes in the copy the main image won't get modified and vice-versa
elizabeth_copy = elizabeth.copy()

elizabeth_copy[::] = 0 # change all pixel colors to 0 -> black
show_image(elizabeth_copy)

print('Memory Address of Elizabeth Image: ', id(elizabeth))
print('Memory Address of Elizabeth Copy Image: ', id(elizabeth_copy))
```

### **Flipping / Mirroring**

```python
# Flipping/Mirroring Images

from PIL import ImageOps

elizabeth_olsen = Image.open('images/elizabeth-olsen.png')

im_flip = ImageOps.flip(elizabeth_olsen)

im_mirror = ImageOps.mirror(elizabeth_olsen)

plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.title("Original")
plt.imshow(elizabeth)
plt.axis("off")

plt.subplot(1, 3, 2)
plt.title("Flipped")
plt.imshow(im_flip)
plt.axis("off")

plt.subplot(1, 3, 3)
plt.title("Mirrored")
plt.imshow(im_mirror)
plt.axis("off")

plt.show()
```

```python
# CV flip methods

im_flip = cv2.flip(elizabeth, 0) # 0 -> vertical, 1 -> horizontal
show_image(im_flip)
```

```python
im_rotate = cv2.rotate(elizabeth, cv2.ROTATE_90_CLOCKWISE)
show_image(im_rotate)
```

### **Cropping**

```python
image = Image.open('images/elizabeth-olsen.png')
image = np.array(image)

upper, lower = 300, 600
cropped_image = image[upper: lower, :, :]
show_image(cropped_image)
```

```python
left, right = 300, 800
cropped_image = image[:, left:right, :]
show_image(cropped_image)
```

```python
cropped_image = image[upper: lower, left: right, :]
show_image(cropped_image)
```

```python
# PIL Image Objects Crop

image = Image.open('images/elizabeth-olsen.png')

image.crop((left, upper, right, lower))
```
