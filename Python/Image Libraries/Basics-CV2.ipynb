```python
image = cv2.imread('images/billie.png')
```

```python
print(f'''
Type: {type(image)}
Shape: {image.shape}
''')
```

```python
import matplotlib.pyplot as plt

plt.axis("off")
plt.imshow(image)
```

But the colors are way off, why is that?

Well OpenCV is **BGR** not **RGB** which makes the ordering shuffeled. (Main difference between `PIL` and `OpenCV`)

```python
new_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.axis("off")
plt.imshow(new_image)
```

```python
# you can convert the image to gray scale using cvtColor
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.axis("off")
plt.imshow(image_gray, cmap="gray")
```

```python
# save the image
cv2.imwrite('images/billie-grayscale.png', image_gray)
```

```python
# we can also load gray scale images with imread
img_gray = cv2.imread("images/billie.png", cv2.IMREAD_GRAYSCALE)
plt.axis("off")
plt.imshow(img_gray, cmap="gray")
```
