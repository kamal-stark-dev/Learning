```python
from PIL import Image
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show_image(image, dpi = 200):
    h, w, _ = image.shape
    plt.figure(figsize=(h / dpi, w / dpi))
    plt.imshow(image)
    plt.axis("off")
```

```python
elizabeth = cv2.imread("images/elizabeth-olsen.png")
elizabeth = cv2.cvtColor(elizabeth, cv2.COLOR_BGR2RGB)

show_image(elizabeth)
```

```python
# if you make changes in the copy the main image won't get modified and vice-versa
elizabeth_copy = elizabeth.copy()

elizabeth_copy[::] = 0 # change all pixel colors to 0 -> black
show_image(elizabeth_copy)

print('Memory Address of Elizabeth Image: ', id(elizabeth))
print('Memory Address of Elizabeth Copy Image: ', id(elizabeth_copy))
```

```python
# Flipping/Mirroring Images

from PIL import ImageOps

elizabeth_olsen = Image.open('images/elizabeth-olsen.png')

im_flip = ImageOps.flip(elizabeth_olsen)

im_mirror = ImageOps.mirror(elizabeth_olsen)

plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.title("Original")
plt.imshow(elizabeth)
plt.axis("off")

plt.subplot(1, 3, 2)
plt.title("Flipped")
plt.imshow(im_flip)
plt.axis("off")

plt.subplot(1, 3, 3)
plt.title("Mirrored")
plt.imshow(im_mirror)
plt.axis("off")

plt.show()
```

```python
# CV flip methods

im_flip = cv2.flip(elizabeth, 0) # 0 -> vertical, 1 -> horizontal
show_image(im_flip)
```

```python
im_rotate = cv2.rotate(elizabeth, cv2.ROTATE_90_CLOCKWISE)
show_image(im_rotate)
```
