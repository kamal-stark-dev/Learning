## Matplotlib

**Matplotlib** is called the "godfather" library for visualization and more advanced libraries for visualization are built on top of this library. It was created by John Hunter.

### Benefits - 
1. High Quality output in multiple formats.
2. Very Customizable.
3. Porivides great control over elements.

Official Link - https://matplotlib.org/

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4, 5, 6, 7, 8])
y = np.array([4, 2, 12, 8, 11, 0, 3, 14])
```

```python
plt.plot(x, y,
          marker="o", mfc="white",
          ls="-.", color="darkgreen"
          )
plt.show()
```

```python
# temperature measures
tempMax = [40, 42, 43, 40, 38, 41, 40]
tempMin = [28, 30, 29, 26, 25, 29, 28]
tempAvg = [35, 36, 34, 35, 36, 34, 35]

# weekdays = list(range(1, 8))
weekdays = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
```

```python
# g*-- is shortcut => `g` is green color, `*` is star marker and `--` is the dashed pattern
plt.plot(weekdays, tempMin, 'g*--', label="Temp - Min")
plt.plot(weekdays, tempAvg, 'bo-.', label="Temp - Avg")
plt.plot(weekdays, tempMax, 'r^--', label="Temp - Max")
plt.legend(bbox_to_anchor=(1, 1))

# giving a title to the whole graph
plt.title("Temperature readings of this week")

# you can customize the fonts of the title and labels - change sizes, colors, fonts, font-weights etc.

# setting starting and ending limit for y-axis
plt.ylim(20, 45)

# labelling the axes
plt.xlabel("Days", color="blue")
plt.ylabel("Temperature", color="red")

# plotting a grid on the graph for better view
plt.grid()

# to save the image you can use the savefig function, use this before the `plt.show()`
plt.tight_layout() # for out of the box legend
# plt.savefig("./figures/Temperature-Readings.png")

plt.show()
```

### Piecharts

```python
revenue = [4.2, 9.4, 38.7, 17.1]
company = ["Mi", "Vivo", "Apple", "Samsung"]

plt.pie(revenue, labels=company, autopct="%0.1F%%",
        explode=[0, 0.1, 0, 0],
        wedgeprops={
          'edgecolor': 'black',
          'lw' : 0.8
        },
        counterclock=False, startangle=80)

# plt.savefig("./figures/Pie-Chart-Revenue.png") # you can save your files in multiple formats

plt.show()
```

### Bar-charts, Histogram, Vertical & Stacked graphs

```python
revenue = [6.2, 9.4, 38.7, 17.1]
profit = [1.9, 2.8, 14.2, 5.8]
company = ["Mi", "Vivo", "Apple", "Samsung"]

xpos = np.arange(len(company))
plt.xticks(xpos, company)

plt.bar(xpos - 0.2, revenue, label="Revenue", width=0.4)
plt.bar(xpos + 0.2, profit, label="Profit", width=0.4)

plt.legend()

# plt.savefig("./figures/Revene-&-Profits-Bar-Graph.png")

plt.show()
```

```python
x_data = np.random.random(1000) * 100
y_data = np.random.random(1000) * 100


plt.title("Randomly scattered 1000 points")
plt.scatter(x_data, y_data, c="#0fea87", marker="s", alpha=0.4)

plt.savefig("./figures/Randomly-Scattered-Plot.png")
plt.show()
```

### Line Plots - for time series data

```python
years = [2005 + y for y in range(20)]
weights = [np.random.randint(60, 85) for _ in range(20)]

plt.plot(years, weights, "go-.") # "go-." => color="green", marker="o", linestyle="dashdot"

plt.xticks(years, rotation=45)
# plt.grid()

plt.title("Random Plot")

# plt.savefig("./figures/Line-Plot-Random.png")

plt.show()
```

### Bar Plots

```python
languages = ["Python", 'C', "C++", 'Java', 'JavaScript', "C#"]
votes = [124, 19, 65, 78, 93, 48]

plt.bar(languages, votes, color="#49fe98", align="center", edgecolor="#000", lw=0.5)

# Meta
plt.title("Programming Languages Votes")
plt.xlabel("Languages")
plt.ylabel("Votes")

# plt.savefig("./figures/Bar-Graph-Votes.png")

plt.show()
```

```python
# below is a normally distributed 1000 valued data with a mean of 20 and standard deviation of 2.5
ages = np.random.normal(20, 2.5, 1000)

plt.title("Histogram of normally distributed ages")
plt.hist(ages, bins=16)
plt.show()

plt.title("Cumulative Graph - Tells the number of people(y-axis) smaller than or equal to age(x-axis)")
plt.hist(ages, bins=16, cumulative=True, color="orange")
# plt.savefig("./figures/Cumulative-Graph.png")
plt.show()
```

### Bos Plot

```python
heights = np.random.normal(170, 8, 500)

plt.boxplot(heights);
```

```python
# Example to get better understanding of boxplot

Q1 = np.linspace(0, 10, 25)
Q2 = np.linspace(10, 100, 25)
Q3 = np.linspace(110, 120, 25)
Q4 = np.linspace(120, 130, 25)

data = np.concatenate((Q1, Q2, Q3, Q4))

plt.boxplot(data);

# in the boxplot you can see that the 1st, 3rd, and 4th quantiles (percentiles) are similar in size as their spread range is 10 and the 2nd one is spreaded by 90 (big)
```

```python
years = [2016 + _ for _ in range(8)]
income = [55, 56, 62, 60, 72, 72, 75, 78]

# editing the ticks values

income_ticks = list(range(50, 81, 2))

plt.plot(years, income, marker="s", ls="-.")

plt.title("Income of John", color="orange", fontname="Agency FB", fontsize=20, fontweight="bold")
plt.xlabel("Year", color="blue")
plt.ylabel("Yearly Income in USD", color="green")

plt.yticks(income_ticks, [f"${x}K" for x in income_ticks])

# plt.savefig("./figures/Personalized-Plot-Ticks-&-Fonts.png")

plt.show()
```

### Style in Matplotlib - 

There are many prebiuit styles in Matplotlib library which you can see [here](https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html) and if you want to make your own style you can visit [here](https://matplotlib.org/stable/users/explain/customizing.html) to know more aobut it.

```python
from matplotlib import style

apple = [100, 102, 99, 101, 101, 100, 102]
nvidia = [91, 90, 102, 104, 105, 102, 109]
tesla = [110, 115, 100, 105, 96, 93, 91]

# Different styles - https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html
style.use("ggplot")
plt.title("Plot using `ggplot` style")

plt.plot(apple, label="Apple")
plt.plot(nvidia, label="Nvidia")
plt.plot(tesla, label="Tesla")

plt.legend(loc="lower center")
plt.plot()
```

### Figures & Subplots

```python
apple = [100, 102, 99, 101, 101, 100, 102]
nvidia = [91, 90, 102, 104, 105, 102, 109]

style.use("default")

# in figure multiple different plots are generated

plt.figure(1)
plt.title("Apple Stocks")
plt.plot(apple, color="red")

plt.figure(2)
plt.title("Nvidia Stocks")
plt.plot(nvidia, color="green")

plt.plot()
```

```python
# in subplots one graph is divided into subplots

x = np.arange(101)

fig, axs = plt.subplots(2, 2, figsize=(10, 8))
plt.subplots_adjust(hspace=0.2)
# plt.tight_layout() # auto adjust plots

axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title("Sine Wave")

axs[0, 1].plot(x, np.cos(x))
axs[0, 1].set_title("Cosine Wave")

axs[1, 0].plot(x, np.tan(x))
axs[1, 0].set_title("Tan Curve")

axs[1, 1].plot(x, np.log(x + 1))
axs[1, 1].set_title("Log Curve")

fig.suptitle("4 Plots", fontsize=20, fontweight="bold", y=1)

# plt.savefig("./figures/4-Plots.png", dpi=200) # increase dpi for higher quality, you can also set transparency=True for no background

plt.show()
```
