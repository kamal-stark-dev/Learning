### Seaborn

A more advanced viasualization library built upon the matplotlib library. You also get prebuilt datasets.

To install seaborn in your systems use - 
`pip install seaborn`

```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
planets = sns.load_dataset("planets")
```

```python
tips
```

```python
sns.scatterplot(x="tip", y="total_bill", data=tips, hue="day", size="size", palette="YlGnBu")
```

```python
sns.histplot(tips['tip'], kde=True, bins=20)
```

```python
sns.displot(tips['tip'], kde=True, bins=20)
```

```python
sns.barplot(x="sex", y="tip", data=tips, hue="sex", palette="YlGnBu")
```

```python
sns.boxplot(x="day", y="tip", data=tips, hue="sex")
```

```python
sns.stripplot(x="day", y="tip", data=tips, hue="sex", dodge=True) # dodge - to separate colours
```

```python
sns.jointplot(x="tip", y="total_bill", data=tips, kind="reg") # reg -> regression line
```

```python
sns.jointplot(x="tip", y="total_bill", data=tips, kind="kde", fill=True)
```

```python
sns.jointplot(x="tip", y="total_bill", data=tips, kind="hex")
```

```python
sns.pairplot(titanic.select_dtypes(["number"]), hue="pclass")
```

```python
numeric_titanic = titanic.select_dtypes("number")
numeric_titanic.corr()
```

```python
sns.heatmap(numeric_titanic.corr(), annot=True, cmap="coolwarm")
```
