### **Video Length and Playback Speed Comparision**

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Use NumPy for cleaner, vectorized calculations
VIDEO_LENGTH = 120
playback_speeds = np.arange(0.1, 4.1, 0.1)  # 0.1 to 4.0
video_durations = VIDEO_LENGTH / playback_speeds

# 2. Setup the plot with a modern style
plt.figure(figsize=(10, 6), dpi=100)
plt.plot(playback_speeds, video_durations, color='#1f77b4', linewidth=2, marker='o', markersize=4, label='Duration')

# 3. Add clear labels and title
plt.title(f'Video Playback Duration ({VIDEO_LENGTH}s Original)', fontsize=14, pad=15)
plt.xlabel('Playback Speed (Multiplier)', fontsize=12)
plt.ylabel('Duration (Seconds)', fontsize=12)

# 4. Enhance readability with a grid and specific ticks
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(np.arange(0, 4.5, 0.5))

# 5. Highlight key data points (optional but helpful)
key_speeds = [0.1, 1.0, 2.0, 4.0]
for speed in key_speeds:
    duration = VIDEO_LENGTH / speed
    plt.annotate(f'{duration}s @ {speed}x ({duration / 60}mins)',
                 xy=(speed, duration),
                 xytext=(speed + 0.2, duration + 40),
                 arrowprops=dict(arrowstyle='->'),)

plt.tight_layout()
plt.show()
```
